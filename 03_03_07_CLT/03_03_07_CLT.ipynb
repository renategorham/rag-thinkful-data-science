```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

###Scenario 1  
Popuation probabilities = 0.2 and 0.5 
Sample sizes = 100 , 100
Sample sizes = 1000 , 1000  
Sample sizes = 20 , 20  

Observe the mean and standard deviation.

Expecations:  
With the increase in sample size I would expect the precision of the mean and standard deviation to increase. And with a decrease in sample size a decrease in precision will result.

```python
n1 = 10
n2 = 10
p1 = 0.2
p2 = 0.5

pop1 = np.random.binomial(n1, p1, 10000)
pop2 = np.random.binomial(n2,p2, 10000) 

plt.hist(pop1, density = True, alpha=0.5, label='Population 1') 
plt.hist(pop2, density = True, alpha=0.5, label='Population 2') 
plt.legend() 
plt.show()
```

```python
sample1 = np.random.choice(pop1, 100, replace = True)
sample2 = np.random.choice(pop2, 100, replace = True)

plt.hist(sample1, density = True, alpha=0.5, label='sample 1') 
plt.hist(sample2, density = True, alpha=0.5, label='sample 2') 
plt.legend() 
plt.show()
```

```python
print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample2.std())
```

```python
sample_means_1 = [np.random.choice(pop1, 100, replace = True).mean() for i in range(25)]
sample_stds_1 = [np.random.choice(pop1, 100, replace = True).std() for i in range(25)]
sample_means_2 = [np.random.choice(pop2, 100, replace = True).mean() for i in range(25)]
sample_stds_2 = [np.random.choice(pop2, 100, replace = True).std() for i in range(25)]

plt.figure(figsize=(15, 8))
plt.subplot(2,2,1)
plt.plot(sample_means_1)
plt.ylim(1.0, 3.0)
plt.axhline(np.mean(sample_means_1), label = np.round(np.mean(sample_means_1),2), color='red',\
                                                      linestyle='solid', linewidth=2)
plt.axhline(np.mean(sample_means_1) - 3*(np.std(sample_means_1)), \
            label = np.round(np.mean(sample_means_1) - 3*(np.std(sample_means_1)),2),\
            color = 'green', linestyle = 'dashed', linewidth = 2)
plt.axhline(np.mean(sample_means_1) + 3*(np.std(sample_means_1)),\
            label = np.round(np.mean(sample_means_1) + 3*(np.std(sample_means_1)),2),\
            color = 'green', linestyle = 'dashed', linewidth = 2)
plt.title('Means - Sample 1')
plt.ylabel('Mean')
plt.legend()

plt.subplot(2,2,2)
plt.plot(sample_stds_1)
plt.ylim(0.5, 2.0)
plt.axhline(np.mean(sample_stds_1), label = np.round(np.mean(sample_stds_1),2), color='red',\
                                                      linestyle='solid', linewidth=2)
plt.axhline(np.mean(sample_stds_1) - 3*(np.std(sample_stds_1)), \
            label = np.round(np.mean(sample_stds_1) - 3*(np.std(sample_stds_1)),2),\
            color = 'green', linestyle = 'dashed', linewidth = 2)
plt.axhline(np.mean(sample_stds_1) + 3*(np.std(sample_stds_1)),\
            label = np.round(np.mean(sample_stds_1) + 3*(np.std(sample_stds_1)),2),\
            color = 'green', linestyle = 'dashed', linewidth = 2)
plt.title('Stds - Sample 1')
plt.ylabel('Standard Deviation')
plt.legend()

plt.subplot(2,2,3)
plt.plot(sample_means_2)
plt.ylim(3.5, 6.5)
plt.axhline(np.mean(sample_means_2), label = np.round(np.mean(sample_means_2),2), color='red',\
                                                      linestyle='solid', linewidth=2)
plt.axhline(np.mean(sample_means_2) - 3*(np.std(sample_means_2)), \
            label = np.round(np.mean(sample_means_2) - 3*(np.std(sample_means_2)),2),\
            color = 'green', linestyle = 'dashed', linewidth = 2)
plt.axhline(np.mean(sample_means_2) + 3*(np.std(sample_means_2)),\
            label = np.round(np.mean(sample_means_2) + 3*(np.std(sample_means_2)),2),\
            color = 'green', linestyle = 'dashed', linewidth = 2)
plt.title('Means - Sample 2')
plt.ylabel('Mean')
plt.legend()

plt.subplot(2,2,4)
plt.plot(sample_stds_2)
plt.ylim(0.5, 2.5)
plt.axhline(np.mean(sample_stds_2), label = np.round(np.mean(sample_stds_2),2), color='red',\
                                                      linestyle='solid', linewidth=2)
plt.axhline(np.mean(sample_stds_2) - 3*(np.std(sample_stds_2)), \
            label = np.round(np.mean(sample_stds_2) - 3*(np.std(sample_stds_2)),2),\
            color = 'green', linestyle = 'dashed', linewidth = 2)
plt.axhline(np.mean(sample_stds_2) + 3*(np.std(sample_stds_2)),\
            label = np.round(np.mean(sample_stds_2) + 3*(np.std(sample_means_2)),2),\
            color = 'green', linestyle = 'dashed', linewidth = 2)
plt.title('Stds - Sample 2')
plt.ylabel('Standard Deviation')
plt.legend()

plt.suptitle('Means and Standard Deviations from Multiple Samples')
plt.show()
```

###Scenario 2  
Popuation probabilities = 0.2 and 0.5 
Popuation probabilities = 0.3 and 0.5  
Popuation probabilities = 0.4 and 0.5  
Sample sizes = 100 , 100

Compute the t-Test.

Expecations:  
As the population probabilities converge the t-statistic will decrease the the p-value will increase.

```python
from scipy.stats import ttest_ind

n1 = 10
n2 = 10
p1 = 0.2
p2 = 0.5

pop1 = np.random.binomial(n1, p1, 10000)
pop2 = np.random.binomial(n2,p2, 10000) 

sample1 = np.random.choice(pop1, 100, replace = True)
sample2 = np.random.choice(pop2, 100, replace = True)

print(ttest_ind(sample2, sample1, equal_var=False))
```

```python
from scipy.stats import ttest_ind

n1 = 10
n2 = 10
p1 = 0.3
p2 = 0.5

pop1 = np.random.binomial(n1, p1, 10000)
pop2 = np.random.binomial(n2,p2, 10000) 

sample1 = np.random.choice(pop1, 100, replace = True)
sample2 = np.random.choice(pop2, 100, replace = True)

print(ttest_ind(sample2, sample1, equal_var=False))
```

t-Statistic has decreased and p-value hase increased.

```python
from scipy.stats import ttest_ind

n1 = 10
n2 = 10
p1 = 0.4
p2 = 0.5

pop1 = np.random.binomial(n1, p1, 10000)
pop2 = np.random.binomial(n2,p2, 10000) 

sample1 = np.random.choice(pop1, 100, replace = True)
sample2 = np.random.choice(pop2, 100, replace = True)

print(ttest_ind(sample2, sample1, equal_var=False))
```

t-Statistic has decreased and p-value has increased.

###Scenario 3  
Distributions from binomial to Poisson.

```python
c1 = 35
c2 = 45


pop1 = np.random.poisson(c1, 10000)
pop2 = np.random.poisson(c2, 10000) 

plt.hist(pop1, density = True, alpha=0.5, label='Population 1') 
plt.hist(pop2, density = True, alpha=0.5, label='Population 2') 
plt.legend() 
plt.show()
```

```python
sample_means_1 = [np.random.choice(pop1, 100, replace = True).mean() for i in range(25)]
sample_means_2 = [np.random.choice(pop2, 100, replace = True).mean() for i in range(25)]

plt.hist(sample_means_1, density = True, alpha=0.5)
plt.axvline(np.mean(sample_means_1), label = np.round(np.mean(sample_means_1),2), color='red',\
                                                      linestyle='solid', linewidth=2)
plt.hist(sample_means_2, density = True, alpha=0.5) 
plt.axvline(np.mean(sample_means_2), label = np.round(np.mean(sample_means_2),2), color='red',\
                                                      linestyle='solid', linewidth=2)
plt.legend() 
plt.show()
```

When changing from a binomial distrbution to a Poisson distribution the sample means still represent the population means. As the population mean for set 1 is 35.0 and the sample mean converges to 35. Likewise the population mean for set 2 is 45.0 and the sample mean converges to 45.
