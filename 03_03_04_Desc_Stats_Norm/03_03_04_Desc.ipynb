```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

###1 - Poisson

```python
__lambda = 6
d1_poisson = np.random.poisson(__lambda, 100)
mean = d1_poisson.mean()
stdev = (d1_poisson.mean())**0.5
plt.hist(d1_poisson, density=True, color = 'lightblue')
plt.axvline(mean, label = np.round(mean,2), color='red', linestyle='solid', linewidth=2)
plt.axvline(mean - stdev, label = np.round(mean - stdev,2), color = 'green', linestyle = 'dashed', linewidth = 2)
plt.axvline(mean + stdev, label = np.round(mean + stdev,2), color = 'green', linestyle = 'dashed', linewidth = 2)
plt.title('Poisson Distribution')
plt.xlabel('Random Data')
plt.ylabel('Density')
plt.legend()
plt.show()
```

The Poisson distribution is the counts per interval, such as, the number of cars passing through a stop light, the number of goals scored per game, and (famously) the number of deaths by horse kicks during battles in the Napoleonic wars. The distribution is composed only by natural numbers and approaches the normal distribution when the mean is greater than 30. Variance is equal to the mean.

###2 - Lognormal

```python
a = 3
b = 0.5

d2_lognormal = np.random.lognormal(a, b, 100)
l = np.mean(np.log(d2_lognormal))
s = np.std(np.log(d2_lognormal))
mean = np.exp(l + s**2/2)
#stdev = np.sqrt(np.exp(2*(l + s**2)) * (np.exp(s**2) - 1))
plt.hist(d2_lognormal, density=True, color = 'lightblue')
plt.axvline(mean, label = np.round(mean,2), color='red', linestyle='solid', linewidth=2)
plt.axvline(mean / s, label = np.round(mean / s,2), color = 'green', linestyle = 'dashed', linewidth = 2)
plt.axvline(mean * s, label = np.round(mean * s,2), color = 'green', linestyle = 'dashed', linewidth = 2)
plt.title('Lognormal Distribution')
plt.xlabel('Random Data')
plt.ylabel('Density')
plt.legend()
plt.show()
```

The lognormal distribution is the distribution of log-transformed normal data. The distribution occurs naturally from processes where output is bounded by zero and the error term is multiplicative, therefore the mean is either multiplied or divided by the standard deviation to set dispersion boundaries.

###3 - Binomial

```python
n = 20
p0 = 0.6

d3_binomial = np.random.binomial(n, p0, 100)
p = np.mean(d3_binomial/n)
mean = n * p
stdev = (n*p*(1-p))**0.5
plt.hist(d3_binomial, density=True, color = 'lightblue')
plt.axvline(mean, label = np.round(mean,2), color='red', linestyle='solid', linewidth=2)
plt.axvline(mean - stdev, label = np.round(mean - stdev,2), color = 'green', linestyle = 'dashed', linewidth = 2)
plt.axvline(mean + stdev, label = np.round(mean + stdev,2), color = 'green', linestyle = 'dashed', linewidth = 2)
plt.title('Binomial Distribution')
plt.xlabel('Random Data')
plt.ylabel('Density')
plt.legend()
plt.show()
```

The binomial distribution is the distribution of successful outcomes in some number of trials given a discrete probability. The mean is the number of trials multiplied by the probability (np) and the variance is the probability of success multiplied by the probability of non-success (p*(1-p)).

###4 - Student's t

```python
df = 10

d4_t = np.random.standard_t(df, 100)+ df
v = d4_t.mean()
stdev = np.sqrt(v / (v-2))
plt.hist(d4_t, density=True, color = 'lightblue')
plt.axvline(v, label = np.round(v,2), color='red', linestyle='solid', linewidth=2)
plt.axvline(v - stdev, label = np.round(v - stdev,2), color = 'green', linestyle = 'dashed', linewidth = 2)
plt.axvline(v + stdev, label = np.round(v + stdev,2), color = 'green', linestyle = 'dashed', linewidth = 2)
plt.title('Student\'s t Distribution')
plt.xlabel('Random Data')
plt.ylabel('Density')
plt.legend()
plt.show()
```

Student's t distribution is a distribution estimating the normal distribution when the sample size is small, and the population standard deviation is unknown. The mean is the degrees of freedom (df), and the variance is degrees of freedom divided by degrees of freedom less 2 (df/df-2).
Numpy produces a standardized distribution, in the chart above it was scaled back to the mean.

###5 - Chi Square

```python
df = 10

d5_chisquare = np.random.chisquare(df, 100)
mean = d5_chisquare.mean()
stdev = np.sqrt(2*mean)
plt.hist(d5_chisquare, density=True, color = 'lightblue')
plt.axvline(mean, label = np.round(mean,2), color='red', linestyle='solid', linewidth=2)
plt.axvline(mean - stdev, label = np.round(mean - stdev,2), color = 'green', linestyle = 'dashed', linewidth = 2)
plt.axvline(mean + stdev, label = np.round(mean + stdev,2), color = 'green', linestyle = 'dashed', linewidth = 2)
plt.title('Chi Square Distribution')
plt.xlabel('Random Data')
plt.ylabel('Density')
plt.legend()
plt.show()
```

The chi squared distribution is the square of the standard normal distribution. The distribution is widely used assess inferential statistics. The mean is the degrees of freedom and the variance is twice the degrees of freedom.

###6 - Beta

```python
a = 3
b = 6

d6_beta = np.random.beta(a, b, 100)
mean = a / (a + b)
stdev = np.sqrt((a * b) / ((a + b)**2 * (a + b + 1)))
plt.hist(d6_beta, density=True, color = 'lightblue')
plt.axvline(mean, label = np.round(mean,2), color='red', linestyle='solid', linewidth=2)
plt.axvline(mean - stdev, label = np.round(mean - stdev,2), color = 'green', linestyle = 'dashed', linewidth = 2)
plt.axvline(mean + stdev, label = np.round(mean + stdev,2), color = 'green', linestyle = 'dashed', linewidth = 2)
plt.title('Beta Distribution')
plt.xlabel('Random Data')
plt.ylabel('Density')
plt.legend()
plt.show()
```

The beta distribution is a continuous distribution between the interval of 0 and 1. The distribution is often used for investigating percentages and proportions. The distribution takes two shape parameters. The shape parameters are used to calculate the mean and variance.

###7 - Combination of two normal distributions

```python
a = 5
b = 0.5
c = 10
d = 1.0

d7_normal = np.random.normal(a, b, 100) + np.random.normal(c, d, 100)
mean = d7_normal.mean()
stdev = d7_normal.std()
plt.hist(d7_normal, density=True, color = 'lightblue')
plt.axvline(mean, label = np.round(mean,2), color='red', linestyle='solid', linewidth=2)
plt.axvline(mean - stdev, label = np.round(mean - stdev,2), color = 'green', linestyle = 'dashed', linewidth = 2)
plt.axvline(mean + stdev, label = np.round(mean + stdev,2), color = 'green', linestyle = 'dashed', linewidth = 2)
plt.title('Normal Distribution')
plt.xlabel('Random Data')
plt.ylabel('Density')
plt.legend()
plt.show()
```

Elements of two random sets were added together (much like a Monte Carlo simulation). Each set was sampled from normal distributions, the first with a mean of 5 and standard deviation of 0.5, the second with a mean of 10 and standard deviation of 1.0. The resulting distribution is (or should be normal) with a mean of 15 (10 + 5) and standard deviation of 1.12 (0.5<sup>2</sup> + 1.0<sup>2</sup>).
